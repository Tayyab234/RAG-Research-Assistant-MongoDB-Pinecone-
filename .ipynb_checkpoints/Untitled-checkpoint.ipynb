```python
from fastapi import FastAPI, UploadFile, File, HTTPException, BackgroundTasks
import os
import uuid

from your_module import process_document

app = FastAPI()

UPLOAD_DIR = "uploads"
os.makedirs(UPLOAD_DIR, exist_ok=True)

MAX_FILE_SIZE_MB = 20


def process_in_background(file_path, filename):
    try:
        result = process_document(file_path, filename)
        print("Processing Done:", result["num_chunks"])
    except Exception as e:
        print("Processing Failed:", str(e))


@app.post("/upload")
async def upload_document(
    background_tasks: BackgroundTasks,
    file: UploadFile = File(...)
):
    file_id = str(uuid.uuid4())
    file_path = os.path.join(UPLOAD_DIR, file_id + "_" + file.filename)

    content = await file.read()

    size_mb = len(content) / (1024 * 1024)
    if size_mb > MAX_FILE_SIZE_MB:
        raise HTTPException(status_code=400, detail="File too large")

    with open(file_path, "wb") as f:
        f.write(content)

    # 🔥 Run in background
    background_tasks.add_task(process_in_background, file_path, file.filename)

    return {
        "file_id": file_id,
        "filename": file.filename,
        "message": "File uploaded. Processing started in background."
    }
```

```python
import os
import tiktoken
from langchain.document_loaders import (
    PyPDFLoader,
    TextLoader,
    UnstructuredFileLoader
)
from langchain.text_splitter import RecursiveCharacterTextSplitter


# -------------------------------
# 1. Select Loader
# -------------------------------
def get_loader(file_path, filename):
    if filename.endswith(".pdf"):
        return PyPDFLoader(file_path)
    elif filename.endswith(".txt"):
        return TextLoader(file_path)
    else:
        return UnstructuredFileLoader(file_path)


# -------------------------------
# 2. Load Documents (Lazy or Full)
# -------------------------------
def load_documents(loader, file_path, lazy_threshold_mb=5):
    file_size_mb = os.path.getsize(file_path) / (1024 * 1024)

    documents = []

    if file_size_mb > lazy_threshold_mb:
        for doc in loader.lazy_load():
            documents.append(doc)
    else:
        documents = loader.load()

    return documents


# -------------------------------
# 3. Token Estimation
# -------------------------------
def estimate_tokens(text):
    enc = tiktoken.get_encoding("cl100k_base")
    return len(enc.encode(text))


# -------------------------------
# 4. Validate Token Limit
# -------------------------------
def validate_tokens(documents, max_tokens=500000):
    total_tokens = 0

    for doc in documents:
        total_tokens += estimate_tokens(doc.page_content)

        if total_tokens > max_tokens:
            raise ValueError("Document exceeds token limit")

    return total_tokens


# -------------------------------
# 5. Clean Documents (Optional but Recommended)
# -------------------------------
def clean_documents(documents):
    cleaned_docs = []

    for doc in documents:
        text = doc.page_content

        # Basic cleaning
        text = text.strip()
        text = " ".join(text.split())  # remove extra spaces

        doc.page_content = text
        cleaned_docs.append(doc)

    return cleaned_docs


# -------------------------------
# 6. Split into Chunks
# -------------------------------
def split_documents(documents, chunk_size=500, chunk_overlap=100):
    splitter = RecursiveCharacterTextSplitter(
        chunk_size=chunk_size,
        chunk_overlap=chunk_overlap
    )

    return splitter.split_documents(documents)


# -------------------------------
# 🔥 MAIN FUNCTION
# -------------------------------
def process_document(file_path, filename):
    # Step 1: Loader
    loader = get_loader(file_path, filename)

    # Step 2: Load
    documents = load_documents(loader, file_path)

    if not documents:
        raise ValueError("No content extracted from document")

    # Step 3: Validate tokens
    total_tokens = validate_tokens(documents)

    # Step 4: Clean
    documents = clean_documents(documents)

    # Step 5: Chunking
    chunks = split_documents(documents)

    return {
        "chunks": chunks,
        "num_chunks": len(chunks),
        "total_tokens": total_tokens
    }
```

```python
import asyncio
from pinecone import Pinecone
import os
from dotenv import load_dotenv

# Load environment variables from .env
load_dotenv()

pc = Pinecone(api_key=os.environ.get("PINECONE_API_KEY"))
desc = pc.describe_index("main-index")
index = pc.Index(host=desc.host)
async def delete_file_vectors(user_id: str, file_id: str):
    """
    Deletes all vectors in a user's namespace that match the given file_id
    """

    namespace = user_id

    try:
        await asyncio.to_thread(
            index.delete,
            namespace=namespace,
            filter={
                "file_id": file_id
            }
        )

        return {
            "status": "success",
            "message": f"Vectors deleted for file_id: {file_id}"
        }

    except Exception as e:
        return {
            "status": "error",
            "message": str(e)
        }
try:
    loop = asyncio.get_running_loop()
    await delete_file_vectors(
        "69d9fc817de60057e7ae93b5",
        "27e7a335-1436-4f86-89d3-fa89719b36fa"
    )
except RuntimeError:
    asyncio.run(
    delete_file_vectors(
        "69d9fc817de60057e7ae93b5",
        "27e7a335-1436-4f86-89d3-fa89719b36fa"
    )
)
```

```python
from fastapi import FastAPI, HTTPException
from pydantic import BaseModel
from datetime import datetime
from typing import List, Optional
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
from Model import deepseek_llm, llm  # LangChain LLMs
from motor.motor_asyncio import AsyncIOMotorClient
import os
import pinecone

# MongoDB setup
MONGO_URI = os.environ.get("MONGO_URI")
mongo_client = AsyncIOMotorClient(MONGO_URI)
db = mongo_client['your_db_name']
user_files_collection = db['user_files']

# Pinecone setup
PINECONE_API_KEY = os.environ.get("PINECONE_API_KEY")
pinecone.init(api_key=PINECONE_API_KEY, environment="us-west1-gcp")  # adjust region
index = pinecone.Index("main-index")

# FastAPI app
app = FastAPI()

# Request body
class QueryRequest(BaseModel):
    query: str
    user_id: str
    model_type: str = "deepseek"  # "deepseek" or "openai"

# Response body
class MultiQueryResponse(BaseModel):
    original_query: str
    generated_queries: List[str]

# Helper to get active files for user
async def get_active_file_text(user_id: str) -> str:
    active_file = await user_files_collection.find_one({"user_id": user_id, "is_active": True})
    if not active_file:
        return ""
    # You can add more: load file content from DB or filesystem
    return active_file.get("filename", "")

# Helper to search Pinecone for context if query is context-specific
async def search_pinecone(user_id: str, query: str) -> List[str]:
    # Embedding generation (example using deepseek_llm or llm)
    embedding = llm.embed_query(query)  # adjust if deepseek_llm has embed method

    # Use Pinecone search with user_id as namespace
    result = index.query(
        vector=embedding,
        top_k=5,
        namespace=user_id,
        include_metadata=True
    )
    # Collect text chunks from metadata
    return [item['metadata']['text'] for item in result['matches']] if result['matches'] else []

# Main endpoint
@app.post("/generate-multiquery", response_model=MultiQueryResponse)
async def generate_multiquery(request: QueryRequest):
    query = request.query
    user_id = request.user_id
    model_type = request.model_type.lower()

    try:
        # Pick LLM
        model = deepseek_llm if model_type == "deepseek" else llm

        # Determine context
        if query.lower().strip() in ["generate q/a", "generate mcqs", "generate questions", "generate answers"]:
            # Generic query → use active file(s) or empty context
            context_text = await get_active_file_text(user_id)
        else:
            # Context-specific → search Pinecone for top matching chunks
            search_results = await search_pinecone(user_id, query)
            context_text = "\n".join(search_results)

            # If no relevant chunks found, fallback to active file
            if not context_text:
                context_text = await get_active_file_text(user_id)

        # Prompt template
        prompt_template = PromptTemplate(
            input_variables=["query", "context"],
            template="Context:\n{context}\n\nGenerate multiple queries for: {query}"
        )

        chain = LLMChain(llm=model, prompt=prompt_template)
        # Invoke the LLM using LangChain
        generated_text = await chain.arun({"query": query, "context": context_text})

        # Split into multiple queries
        generated_queries = [q.strip() for q in generated_text.split("\n") if q.strip()]

        return MultiQueryResponse(
            original_query=query,
            generated_queries=generated_queries
        )

    except Exception as e:
        raise HTTPException(status_code=500, detail=str(e))
```
